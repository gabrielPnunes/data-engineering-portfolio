```python
from extractor import extract_all_files, inspect, check_duplicates, extract_summary
from storage import save_csv, save_parquet
```

### CAMINHOS E CONFIGS

```python
RAW_PATH       = r"C:/Users/Ceifas/OneDrive\Desktop\data-engineering-portfolio/projects/month02_Pipeline/Data/Raw"
PROCESSED_PATH = r"C:/Users/Ceifas/OneDrive\Desktop\data-engineering-portfolio/projects/month02_Pipeline/Data/Processed"
```

```python
FILES = {
    "deliveries": {
        "file":   "deliveries.csv",
        "schema": ["delivery_id", "driver_id", "hub_id", "status",
                   "shipped_date", "delivered_date", "distance_km",
                   "cost", "customer_rating"],
        "params": {"sep": ",", "encoding": "utf-8"}
    },
    "drivers": {
        "file":   "drivers.csv",
        "schema": ["driver_id", "name", "hub_id"],
        "params": {"sep": ",", "encoding": "utf-8"}
    },
    "hubs": {
        "file":   "hubs.csv",
        "schema": ["hub_id", "city", "state"],
        "params": {"sep": ",", "encoding": "utf-8"}
    },
}
```

### EXTRACT

```python
print("\n===EXTRACT===")
dfs = extract_all_files(RAW_PATH, FILES)
```

```python
print("\n===INSPECIONAR===")
if "deliveries" in dfs:
    inspect(dfs["deliveries"], "deliveries")
```

```python
if "drivers" in dfs:
    inspect(dfs["drivers"], "drivers")
```

```python
if "hubs" in dfs:
    inspect(dfs["hubs"], "hubs")
```

```python
print("\n===VALIDAR DUPLICATAS===")
check_duplicates(dfs["deliveries"], "delivery_id", "deliveries")
check_duplicates(dfs["drivers"],    "driver_id",   "drivers")
check_duplicates(dfs["hubs"],       "hub_id",      "hubs")
```

```python
print("\n===RESUMO===")
df_summary = extract_summary(dfs)
print(df_summary.to_string(index=False))
```

```python
print("\n===SAVE RAW===")
for name, df in dfs.items():
    save_parquet(df, f"{PROCESSED_PATH}\\raw_{name}.parquet")
    save_csv(df,     f"{PROCESSED_PATH}\\raw_{name}.csv")

save_csv(df_summary, f"{PROCESSED_PATH}\\extract_summary.csv")
```
