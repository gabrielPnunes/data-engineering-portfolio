```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["estudos"]
colecao = db["usuarios"]
```

```python
colecao.insert_many([
    {"nome": "Ana", "idade": 28, "cidade": "SP"},
    {"nome": "Bruno", "idade": 35, "cidade": "RJ"},
    {"nome": "Carlos", "idade": 28, "cidade": "SP"}
])
```

```python
list(colecao.find({"cidade": "SP"}))
```

```python
pipeline = [
    {"$group": {"_id": "$cidade", "total": {"$sum": 1}}}
]

list(colecao.aggregate(pipeline))
```
