## Tipos Básicos

```python
#type() é uma função importante para identificar o tipo da váriavel
numero_aleatorio = randint(0, 15)
print(f"Variável 'numero_aleatorio': {numero_aleatorio}, Tipo: {type(numero_aleatorio)}")
print("-" * 30)

# INT, são tipos inteiros geralmente usados para ID's, quatidade de registros.
tipo_int = 20 # mesmo não declarando ela ja vem como inteiro isso acontece com todos os tipos pois ela é de tipagem dinâmica
print(f"Variável 'tipo_int' (original): {tipo_int}, Tipo: {type(tipo_int)}")
# int(tipo_int) # Isso não altera a variável 'tipo_int', apenas retorna um inteiro
print(f"Variável 'tipo_int' (convertida para int): {int(tipo_int)}, Tipo: {type(int(tipo_int))}")
print("-" * 30)

# FLOAT; são numeros quebrados; como médias, numeros financeiros, métricas(kg's, Altura etc).
tipo_float = 13.5
print(f"Variável 'tipo_float': {tipo_float}, Tipo: {type(tipo_float)}")
# float(tipo_float) # Isso não altera a variável 'tipo_float', apenas retorna um float
print(f"Variável 'tipo_float' (convertida para float): {float(tipo_float)}, Tipo: {type(float(tipo_float))}")
print("-" * 30)

# STR; são cadeias de texto; nomes, mensagem, email, telefone, cpf.
tipo_str = "Bom Dia!!"
print(f"Variável 'tipo_str': {tipo_str}, Tipo: {type(tipo_str)}")
# str(tipo_str) # Isso não altera a variável 'tipo_str', apenas retorna uma string
print(f"Variável 'tipo_str' (convertida para str): {str(tipo_str)}, Tipo: {type(str(tipo_str))}")
print("-" * 30)

# CONVERSÃO DE TIPOS
print("--- Conversão de Tipos ---")
conversao_int = int(tipo_float)
print(f"Conversão de float para int: {conversao_int}, Tipo: {type(conversao_int)}")

conversao_flaot = float(tipo_int)
print(f"Conversão de int para float: {conversao_flaot}, Tipo: {type(conversao_flaot)}")

conversao_str = str(tipo_int)
print(f"Conversão de int para str: {conversao_str}, Tipo: {type(conversao_str)}")
print("-" * 30)
```

## Operadores

#### Operadores Aritméticos

```python
# Operadores Aritméticos

# Soma +; pode ser feita por diferentes tipos numericos; pode ser feita usando variaveis;
soma_1 = 59 + 58
print(f"Soma (59 + 58): {soma_1}")# esse padrão de print é muito util pois formata uma string e usando as {chaves} usamos qualquer expressão

soma_2 = numero_aleatorio + 9
print(f"Soma (numero_aleatorio + 9): {soma_2}")

soma_3 = 15 + 5.38
print(f"Soma (15 + 5.38): {soma_3}")
print("-" * 20) #isso é bacisamente a strings concatenada 20 vezes

# Subtração -; pode ser feita por diferentes tipos numericos; pode ser feita usando variaveis; mesma coisa da soma
subtracao_1 = 50 - 3
print(f"Subtração (50 - 3): {subtracao_1}")

subtracao_2 = 90 - 99.5
print(f"Subtração (90 - 99.5): {subtracao_2}")
subtracao_3 = numero_aleatorio - numero_aleatorio
print(f"Subtração (numero_aleatorio - numero_aleatorio): {subtracao_3}")
print("-" * 20)

# Multiplicação *; mesmo esquema das outras
multplicacao_1 = 50 * 3
print(f"Multiplicação (50 * 3): {multplicacao_1}")

multiplicao_2 = 0.3 * -0.589
print(f"Multiplicação (0.3 * -0.589): {multiplicao_2}")

multiplicao_3 = numero_aleatorio * 8
print(f"Multiplicação (numero_aleatorio * 8): {multiplicao_3}")
print("-" * 20)

# Exponencial **; ou pow(base, expoente)
exponencial_1 = 10 ** 3
print(f"Exponencial (10 ** 3): {exponencial_1}")

exponencial_2 = 10 ** -0.1
print(f"Exponencial (10 ** -0.1): {exponencial_2}")

exponencial_3 = numero_aleatorio ** 2
print(f"Exponencial (numero_aleatorio ** 2): {exponencial_3}")
print("-" * 20)

# Divisão /; mesmo esquema das outras, divisao por zero ocorre erro pois não existe
divisao_1 = 50/2
print(f"Divisão (50 / 2): {divisao_1}")

divisao_2 = 500/0.0000000000000000000000000001
print(f"Divisão (500 / 0.0...01): {divisao_2}")
divisao_3 = 5053.50 / numero_aleatorio
print(f"Divisão (5053.50 / numero_aleatorio): {divisao_3}")
print("-" * 20)


# Divisão inteira //; mesma coisa da divisão comum, regras e etc. porem retorna apenas a parte inteira e não UM INTEIRO(INT)
divisao_1_inteira = 51 // 2
print(f"Divisão Inteira (51 // 2): {divisao_1_inteira}")

divisao_2_inteira = 0.15 // 2
print(f"Divisão Inteira (0.15 // 2): {divisao_2_inteira}")

divisao_3_inteira = numero_aleatorio // 5
print(f"Divisão Inteira (numero_aleatorio // 5): {divisao_3_inteira}")
print("-" * 20)


# Resto de divisão %
resto_1 = 10 % 3
print(f"Resto da Divisão (10 % 3): {resto_1}")

resto_2 = 15.5 % 4
print(f"Resto da Divisão (15.5 % 4): {resto_2}")

resto_3 = numero_aleatorio % 10
print(f"Resto da Divisão (numero_aleatorio % 10): {resto_3}")
print("-" * 20)
```

#### Operadores Relacionais ou De Comparação

```python
# Basicamente são todos os operadoes de comparação
comparação_1 = 59
comparação_2 = 23

# IGUAL A (==); retorna true ou false
print(f"{comparação_1} == {comparação_2}: {comparação_1 == comparação_2}")
print("-" * 20)

# DIFERENTE DE (!=); retorna true ou false; porem pode ser usado para comparar DF's ou pegar algo que é diferente disso ou o contrario de algo
print(f"{comparação_1} != {comparação_2}: {comparação_1 != comparação_2}")
print("-" * 20)

# MAIOR QUE (>); retorna true ou false; muito usado para comparar tipos numericos e fazer laços de decisão e repetição
print(f"{comparação_1} > {comparação_2}: {comparação_1 > comparação_2}")
print("-" * 20)

# MAIOR OU IGUAL QUE (>=); retorna true ou false; muito usado para comparar tipos numericos e fazer laços de decisão e repetição
print(f"{comparação_1} >= {comparação_2}: {comparação_1 >= comparação_2}")
print("-" * 20)

# MENOR QUE (<); retorna true ou false; muito usado para comparar tipos numericos e fazer laços de decisão e repetição
print(f"{comparação_1} < {comparação_2}: {comparação_1 < comparação_2}")
print("-" * 20)

# MENOR OU IGUAL QUE (<=); muito usado para comparar tipos numericos e fazer laços de decisão e repetição
print(f"{comparação_1} <= {comparação_2}: {comparação_1 <= comparação_2}")
```

#### Operadores Lógicos

```python
# Os operadores lógicos funcionam para comparar parametros em python

# AND(E) Funciona da seguiente forma, isso E isso tem que acontecer para ir pra frente
idade = 18
tem_carteira = True

print(f"Idade: {idade}, Tem carteira: {tem_carteira}")
if idade >= 18 and tem_carteira:
  print(f"Condição (idade >= 18 AND tem_carteira): {idade >= 18 and tem_carteira} -> habilitado")
else:
  print(f"Condição (idade >= 18 AND tem_carteira): {idade >= 18 and tem_carteira} -> Não Habilitado")
print("-" * 20)

# OR(OU) Funciona da seguiente forma, isso ou isso tem que acontecer pra ir pra frente, não funciona se os dois forem falsos
dia_semana = "sexta"
feriado = False
print(f"Dia da semana: {dia_semana}, Feriado: {feriado}")

if dia_semana == "sabado" or dia_semana == "domingo" or feriado == True:
  print(f"Condição (dia_semana é sabado OU domingo OU feriado): {dia_semana == 'sabado' or dia_semana == 'domingo' or feriado == True} -> folga")
else:
  print(f"Condição (dia_semana é sabado OU domingo OU feriado): {dia_semana == 'sabado' or dia_semana == 'domingo' or feriado == True} -> trabalhar")
  print("-" * 20)

# NOT(NÃO) Funciona da seguiente forma; Ele é um inverso de boleano, o que é false vira verdadeiro o que é verdadeiro vira falso
num1 = 15
num2 = 50

print(f"Expressão ({num1} > {num2}): {num1 > num2}")
print(f"NOT ({num1} > {num2}): {not(num1 > num2)}")
```

## Estruturas de Decisão

```python
from random import randint
#IF(SE); ELSE(SE NÃO); ELIF(SE NÃO SE) são estrutras de decisão que usam o seguinte formato
media = 6
nota = randint(0, 10)

if nota >= media:
  print(f"Sua nota é {nota}, e é classificada como Acima ou Igual à Média")
elif nota > 8.5:
  print(f"Sua nota é {nota}, e é classificada como Muito Acimada da media")
else:
  print(f"Sua nota é {nota}, e é classificada como Abaixo da Média")
```

##Funções Embutidas

```python
# São Basicamente funções que ja vem por padrão no python, sem precisar de import

# len(); retorna o tamanho ou contagem de uma certa lista, df e etc
lista_numeros = [10, 20, 30, 40, 50]
print("Lista de números:", lista_numeros)
print(f"Tamanho da lista (len): {len(lista_numeros)}")
print("-" * 30)

# sum(); soma a estrtura
print(f"Soma dos valores (sum): {sum(lista_numeros)}")
print("-" * 30)

# max() e min() maior e menor valor
print(f"Maior valor (max): {max(lista_numeros)}")
print(f"Menor valor (min): {min(lista_numeros)}")
print("-" * 30)

# abs() remove o negativo de qualquer numero e sempre volta positivo
numero_negativo = -42
print(f"Valor absoluto de {numero_negativo} (abs): {abs(numero_negativo)}")
print("-" * 30)

# round() arrendendo para quantas casas decimais desejar
valor = 12.56789
print(f"Valor original: {valor}")
print(f"Arredondado para 2 casas (round): {round(valor, 2)}")
print("-" * 30)

# sorted() ordena uma estrutura de forma aleatorioa
numeros_desordenados = [5, 2, 9, 1, 7]
print("Lista original:", numeros_desordenados)
print("Lista ordenada (sorted):", sorted(numeros_desordenados))
print("-" * 30)

# range() gera uma estrtura de numeros ou uma lista de x a x
print("Usando range() no for")
for i in range(1, 6):
    print(f"Valor do range: {i}")
print("-" * 30)

# input() recebe dados do usuário (sempre retorna string), isso pode ser convertido
nome_usuario = input("Digite seu nome: ")
print(f"Nome digitado: {nome_usuario}")
print(type(nome_usuario))
```
