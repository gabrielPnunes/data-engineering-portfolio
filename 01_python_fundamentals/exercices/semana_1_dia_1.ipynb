```python
#type() √© uma fun√ß√£o importante para identificar o tipo da v√°riavel
numero_aleatorio = randint(0, 15)
print(f"Vari√°vel 'numero_aleatorio': {numero_aleatorio}, Tipo: {type(numero_aleatorio)}")
print("-" * 30)

# INT, s√£o tipos inteiros geralmente usados para ID's, quatidade de registros.
tipo_int = 20 # mesmo n√£o declarando ela ja vem como inteiro isso acontece com todos os tipos pois ela √© de tipagem din√¢mica
print(f"Vari√°vel 'tipo_int' (original): {tipo_int}, Tipo: {type(tipo_int)}")
# int(tipo_int) # Isso n√£o altera a vari√°vel 'tipo_int', apenas retorna um inteiro
print(f"Vari√°vel 'tipo_int' (convertida para int): {int(tipo_int)}, Tipo: {type(int(tipo_int))}")
print("-" * 30)

# FLOAT; s√£o numeros quebrados; como m√©dias, numeros financeiros, m√©tricas(kg's, Altura etc).
tipo_float = 13.5
print(f"Vari√°vel 'tipo_float': {tipo_float}, Tipo: {type(tipo_float)}")
# float(tipo_float) # Isso n√£o altera a vari√°vel 'tipo_float', apenas retorna um float
print(f"Vari√°vel 'tipo_float' (convertida para float): {float(tipo_float)}, Tipo: {type(float(tipo_float))}")
print("-" * 30)

# STR; s√£o cadeias de texto; nomes, mensagem, email, telefone, cpf.
tipo_str = "Bom Dia!!"
print(f"Vari√°vel 'tipo_str': {tipo_str}, Tipo: {type(tipo_str)}")
# str(tipo_str) # Isso n√£o altera a vari√°vel 'tipo_str', apenas retorna uma string
print(f"Vari√°vel 'tipo_str' (convertida para str): {str(tipo_str)}, Tipo: {type(str(tipo_str))}")
print("-" * 30)

# CONVERS√ÉO DE TIPOS
print("--- Convers√£o de Tipos ---")
conversao_int = int(tipo_float)
print(f"Convers√£o de float para int: {conversao_int}, Tipo: {type(conversao_int)}")

conversao_flaot = float(tipo_int)
print(f"Convers√£o de int para float: {conversao_flaot}, Tipo: {type(conversao_flaot)}")

conversao_str = str(tipo_int)
print(f"Convers√£o de int para str: {conversao_str}, Tipo: {type(conversao_str)}")
print("-" * 30)
```

## Operadores

#### Operadores Aritm√©ticos

```python
# Operadores Aritm√©ticos

# Soma +; pode ser feita por diferentes tipos numericos; pode ser feita usando variaveis;
soma_1 = 59 + 58
print(f"Soma (59 + 58): {soma_1}")# esse padr√£o de print √© muito util pois formata uma string e usando as {chaves} usamos qualquer express√£o

soma_2 = numero_aleatorio + 9
print(f"Soma (numero_aleatorio + 9): {soma_2}")

soma_3 = 15 + 5.38
print(f"Soma (15 + 5.38): {soma_3}")
print("-" * 20) #isso √© bacisamente a strings concatenada 20 vezes

# Subtra√ß√£o -; pode ser feita por diferentes tipos numericos; pode ser feita usando variaveis; mesma coisa da soma
subtracao_1 = 50 - 3
print(f"Subtra√ß√£o (50 - 3): {subtracao_1}")

subtracao_2 = 90 - 99.5
print(f"Subtra√ß√£o (90 - 99.5): {subtracao_2}")
subtracao_3 = numero_aleatorio - numero_aleatorio
print(f"Subtra√ß√£o (numero_aleatorio - numero_aleatorio): {subtracao_3}")
print("-" * 20)

# Multiplica√ß√£o *; mesmo esquema das outras
multplicacao_1 = 50 * 3
print(f"Multiplica√ß√£o (50 * 3): {multplicacao_1}")

multiplicao_2 = 0.3 * -0.589
print(f"Multiplica√ß√£o (0.3 * -0.589): {multiplicao_2}")

multiplicao_3 = numero_aleatorio * 8
print(f"Multiplica√ß√£o (numero_aleatorio * 8): {multiplicao_3}")
print("-" * 20)

# Exponencial **; ou pow(base, expoente)
exponencial_1 = 10 ** 3
print(f"Exponencial (10 ** 3): {exponencial_1}")

exponencial_2 = 10 ** -0.1
print(f"Exponencial (10 ** -0.1): {exponencial_2}")

exponencial_3 = numero_aleatorio ** 2
print(f"Exponencial (numero_aleatorio ** 2): {exponencial_3}")
print("-" * 20)

# Divis√£o /; mesmo esquema das outras, divisao por zero ocorre erro pois n√£o existe
divisao_1 = 50/2
print(f"Divis√£o (50 / 2): {divisao_1}")

divisao_2 = 500/0.0000000000000000000000000001
print(f"Divis√£o (500 / 0.0...01): {divisao_2}")
divisao_3 = 5053.50 / numero_aleatorio
print(f"Divis√£o (5053.50 / numero_aleatorio): {divisao_3}")
print("-" * 20)


# Divis√£o inteira //; mesma coisa da divis√£o comum, regras e etc. porem retorna apenas a parte inteira e n√£o UM INTEIRO(INT)
divisao_1_inteira = 51 // 2
print(f"Divis√£o Inteira (51 // 2): {divisao_1_inteira}")

divisao_2_inteira = 0.15 // 2
print(f"Divis√£o Inteira (0.15 // 2): {divisao_2_inteira}")

divisao_3_inteira = numero_aleatorio // 5
print(f"Divis√£o Inteira (numero_aleatorio // 5): {divisao_3_inteira}")
print("-" * 20)


# Resto de divis√£o %
resto_1 = 10 % 3
print(f"Resto da Divis√£o (10 % 3): {resto_1}")

resto_2 = 15.5 % 4
print(f"Resto da Divis√£o (15.5 % 4): {resto_2}")

resto_3 = numero_aleatorio % 10
print(f"Resto da Divis√£o (numero_aleatorio % 10): {resto_3}")
print("-" * 20)
```

#### Operadores Relacionais ou De Compara√ß√£o

```python
# Basicamente s√£o todos os operadoes de compara√ß√£o
compara√ß√£o_1 = 59
compara√ß√£o_2 = 23

# IGUAL A (==); retorna true ou false
print(f"{compara√ß√£o_1} == {compara√ß√£o_2}: {compara√ß√£o_1 == compara√ß√£o_2}")
print("-" * 20)

# DIFERENTE DE (!=); retorna true ou false; porem pode ser usado para comparar DF's ou pegar algo que √© diferente disso ou o contrario de algo
print(f"{compara√ß√£o_1} != {compara√ß√£o_2}: {compara√ß√£o_1 != compara√ß√£o_2}")
print("-" * 20)

# MAIOR QUE (>); retorna true ou false; muito usado para comparar tipos numericos e fazer la√ßos de decis√£o e repeti√ß√£o
print(f"{compara√ß√£o_1} > {compara√ß√£o_2}: {compara√ß√£o_1 > compara√ß√£o_2}")
print("-" * 20)

# MAIOR OU IGUAL QUE (>=); retorna true ou false; muito usado para comparar tipos numericos e fazer la√ßos de decis√£o e repeti√ß√£o
print(f"{compara√ß√£o_1} >= {compara√ß√£o_2}: {compara√ß√£o_1 >= compara√ß√£o_2}")
print("-" * 20)

# MENOR QUE (<); retorna true ou false; muito usado para comparar tipos numericos e fazer la√ßos de decis√£o e repeti√ß√£o
print(f"{compara√ß√£o_1} < {compara√ß√£o_2}: {compara√ß√£o_1 < compara√ß√£o_2}")
print("-" * 20)

# MENOR OU IGUAL QUE (<=); muito usado para comparar tipos numericos e fazer la√ßos de decis√£o e repeti√ß√£o
print(f"{compara√ß√£o_1} <= {compara√ß√£o_2}: {compara√ß√£o_1 <= compara√ß√£o_2}")
```

#### Operadores L√≥gicos

```python
# Os operadores l√≥gicos funcionam para comparar parametros em python

# AND(E) Funciona da seguiente forma, isso E isso tem que acontecer para ir pra frente
idade = 18
tem_carteira = True

print(f"Idade: {idade}, Tem carteira: {tem_carteira}")
if idade >= 18 and tem_carteira:
  print(f"Condi√ß√£o (idade >= 18 AND tem_carteira): {idade >= 18 and tem_carteira} -> habilitado")
else:
  print(f"Condi√ß√£o (idade >= 18 AND tem_carteira): {idade >= 18 and tem_carteira} -> N√£o Habilitado")
print("-" * 20)

# OR(OU) Funciona da seguiente forma, isso ou isso tem que acontecer pra ir pra frente, n√£o funciona se os dois forem falsos
dia_semana = "sexta"
feriado = False
print(f"Dia da semana: {dia_semana}, Feriado: {feriado}")

if dia_semana == "sabado" or dia_semana == "domingo" or feriado == True:
  print(f"Condi√ß√£o (dia_semana √© sabado OU domingo OU feriado): {dia_semana == 'sabado' or dia_semana == 'domingo' or feriado == True} -> folga")
else:
  print(f"Condi√ß√£o (dia_semana √© sabado OU domingo OU feriado): {dia_semana == 'sabado' or dia_semana == 'domingo' or feriado == True} -> trabalhar")
  print("-" * 20)

# NOT(N√ÉO) Funciona da seguiente forma; Ele √© um inverso de boleano, o que √© false vira verdadeiro o que √© verdadeiro vira falso
num1 = 15
num2 = 50

print(f"Express√£o ({num1} > {num2}): {num1 > num2}")
print(f"NOT ({num1} > {num2}): {not(num1 > num2)}")
```

## Estruturas de Decis√£o

```python
from random import randint
#IF(SE); ELSE(SE N√ÉO); ELIF(SE N√ÉO SE) s√£o estrutras de decis√£o que usam o seguinte formato
media = 6
nota = randint(0, 10)

if nota >= media:
  print(f"Sua nota √© {nota}, e √© classificada como Acima ou Igual √† M√©dia")
elif nota > 8.5:
  print(f"Sua nota √© {nota}, e √© classificada como Muito Acimada da media")
else:
  print(f"Sua nota √© {nota}, e √© classificada como Abaixo da M√©dia")
```

##Fun√ß√µes Embutidas

```python
# S√£o Basicamente fun√ß√µes que ja vem por padr√£o no python, sem precisar de import

# len(); retorna o tamanho ou contagem de uma certa lista, df e etc
lista_numeros = [10, 20, 30, 40, 50]
print("Lista de n√∫meros:", lista_numeros)
print(f"Tamanho da lista (len): {len(lista_numeros)}")
print("-" * 30)

# sum(); soma a estrtura
print(f"Soma dos valores (sum): {sum(lista_numeros)}")
print("-" * 30)

# max() e min() maior e menor valor
print(f"Maior valor (max): {max(lista_numeros)}")
print(f"Menor valor (min): {min(lista_numeros)}")
print("-" * 30)

# abs() remove o negativo de qualquer numero e sempre volta positivo
numero_negativo = -42
print(f"Valor absoluto de {numero_negativo} (abs): {abs(numero_negativo)}")
print("-" * 30)

# round() arrendendo para quantas casas decimais desejar
valor = 12.56789
print(f"Valor original: {valor}")
print(f"Arredondado para 2 casas (round): {round(valor, 2)}")
print("-" * 30)

# sorted() ordena uma estrutura de forma aleatorioa
numeros_desordenados = [5, 2, 9, 1, 7]
print("Lista original:", numeros_desordenados)
print("Lista ordenada (sorted):", sorted(numeros_desordenados))
print("-" * 30)

# range() gera uma estrtura de numeros ou uma lista de x a x
print("Usando range() no for")
for i in range(1, 6):
    print(f"Valor do range: {i}")
print("-" * 30)

# input() recebe dados do usu√°rio (sempre retorna string), isso pode ser convertido
nome_usuario = input("Digite seu nome: ")
print(f"Nome digitado: {nome_usuario}")
print(type(nome_usuario))
```
