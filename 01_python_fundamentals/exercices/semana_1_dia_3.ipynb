```python
#No código anterior eu usava bastante isso, que basicamente gera uma linha, irei transformar em uma função para ficar chamando
#1 - 
print("-" * 30)

def separador():
    print("=" * 40)

separador()


#criar algumas funções numericas para testas
#2 - 
def operacoes(num_1, num_2):
    return num_1 + num_2, num_1 - num_2, num_1 * num_2

soma, sub, mult = operacoes(10,5)# é interresante perceber como dei a cada função os valores todos de uma vezes, se não fosse uma função deveria fazer isso todas as vezes
print(soma, sub, mult)

separador()# a utilidade do separador se mostrando

#criar uma função de de quadrado para usar futuramente no util
#3 -
def num_quadrado(num_1, num_2):
    return num_1 ** num_2
print(num_quadrado(5,3))
```

## Funções

```python
# São blocos de código que executa uma tarefa específica, e é reutilizado varias vezes ao longo do codigo
def mostrar_mensagem(): #Estrutura Base
    print("Oi")

mostrar_mensagem() #Chamando a função

separador()

# Destar forma a função fica fixa e não tem como fazer calculos por exemplos
# Pois ela não não recebe valores externos
def calculo_fixo(num_fixo_1, num_fixo_2):
    print(num_fixo_1 + num_fixo_2)
print(calculo_fixo(5,3)) # aqui se gera um nome pois não existe um return para SALVAR O RESULTADO DA função

#Com Return
def calculo(num_1, num_2):
    return num_1 + num_2

print(calculo(5,3))

separador()
```
