```python
import json

with open("dados.json", "r") as f:
    dados = json.load(f)


with open("dados.json", "w") as f:
    json.dump(dados, f, indent=4)
```

```python
import csv

with open("dados.csv") as f:
    reader = csv.reader(f)
    for linha in reader:
        print(linha)


with open("dados.csv") as f:
    reader = csv.DictReader(f)
    for linha in reader:
        print(linha["nome"])
```
