```python
#Exercício 1 
#Crie uma função validar_idade(idade) que:

#Receba uma variável idade

#Verifique:

#Se não for inteiro, retorne "Idade inválida"
#Se for menor que 18, retorne "Menor de idade"
#Se for entre 18 e 65, retorne "Adulto"
#Se for maior que 65, retorne "Idoso"

def validar_idade(idade):
    try:
        idade = int(idade)
    except ValueError:
        return "Idade inválida"

    if idade < 18:
        return "Menor de idade"
    elif idade <= 65:
        return "Adulto"
    else:
        return "Idoso"


idade = input("Digite sua idade: ")
print(validar_idade(idade))
```
